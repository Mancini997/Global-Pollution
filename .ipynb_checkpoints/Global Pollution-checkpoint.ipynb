```python
import pandas as pd

# Load Air Pollution data

air_pollution_df = pd.read_csv('pollution.csv')

# Load Population data
population_df = pd.read_csv('population.csv')

# Display the first few rows of each dataframe
print(air_pollution_df.head())
print(population_df.head())
#Understand the Structure of the Datasets
```

```python
population_df.head(1)
```

```python
air_pollution_df.info()
```

```python

# Extract unique country names from both datasets
pollution_countries = set(air_pollution_df['Country'].unique())
population_countries = set(population_df['Country'].unique())

# Find the differences between the two sets of country names
countries_in_pollution_not_population = pollution_countries - population_countries
countries_in_population_not_pollution = population_countries - pollution_countries

# Display the differences
print("\nCountries in Air Pollution data but not in Population data:")
print(countries_in_pollution_not_population)

print("\nCountries in Population data but not in Air Pollution data:")
print(countries_in_population_not_pollution)

# Mapping for different country names
mapping = {
    'Russian Federation': 'Russia',
    'United States of America': 'United States',
    'Republic of North Macedonia': 'North Macedonia',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'United Republic of Tanzania': 'Tanzania',
    'Viet Nam': 'Vietnam',
    'Iran (Islamic Republic of)': 'Iran',
    'Democratic Republic of the Congo': 'DR Congo',
    'Czechia': 'Czech Republic (Czechia)',
    "Côte d'Ivoire": "Côte d'Ivoire",
    'Republic of Korea': 'South Korea',
    'Syrian Arab Republic': 'Syria',
    "Lao People's Democratic Republic": 'Laos',
    'Bolivia (Plurinational State of)': 'Bolivia',
    'Venezuela (Bolivarian Republic of)': 'Venezuela',
    'State of Palestine': 'Palestine',
    'Myanmar': 'Burma',  # Assuming Myanmar is referred to as Burma in one of the datasets
    'Eswatini': 'Swaziland',  # Assuming Eswatini is referred to as Swaziland in one of the datasets
}

# Apply the mapping to align the country names
air_pollution_df['Country'] = air_pollution_df['Country'].replace(mapping)
population_df['Country'] = population_df['Country'].replace(mapping)

# Recheck unique country names after mapping
pollution_countries = set(air_pollution_df['Country'].unique())
population_countries = set(population_df['Country'].unique())

# Find the differences again after applying the mapping
countries_in_pollution_not_population = pollution_countries - population_countries
countries_in_population_not_pollution = population_countries - pollution_countries

# Display the differences after mapping
print("\nCountries in Air Pollution data but not in Population data after mapping:")
print(countries_in_pollution_not_population)

print("\nCountries in Population data but not in Air Pollution data after mapping:")
print(countries_in_population_not_pollution)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'merged_dataset' is your DataFrame
# Drop duplicates by country, keeping the maximum PM2.5 AQI Value
merged_dataset_unique = merged_dataset.loc[merged_dataset.groupby('Country')['PM2.5 AQI Value'].idxmax()]

# Verify the unique countries
print(merged_dataset_unique['Country'].value_counts())

# Filter for Top 10 Highest Population Density
top_population_density = merged_dataset_unique.sort_values(by='Density(P/Km²)', ascending=False).head(10)

# Filter for Top 10 Highest PM2.5 Levels
top_pm25 = merged_dataset_unique.sort_values(by='PM2.5 AQI Value', ascending=False).head(10)

# Scatter Plot for Top 10 Highest Population Density vs PM2.5
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Density(P/Km²)', y='PM2.5 AQI Value', data=top_population_density, hue='Country')
sns.regplot(x='Density(P/Km²)', y='PM2.5 AQI Value', data=top_population_density, scatter=False, color='red')
plt.title('Top 10 Highest Population Density vs PM2.5 AQI Value')
plt.xlabel('Population Density (people per km²)')
plt.ylabel('PM2.5 AQI Value')
plt.show()

# Bar Plot for Top 10 Highest PM2.5 Levels
plt.figure(figsize=(10, 6))
sns.barplot(x='PM2.5 AQI Value', y='Country', data=top_pm25, palette='viridis')
plt.title('Top 10 Countries with Highest PM2.5 AQI Value')
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Country')
plt.show()

# Correlation for Top 10 Highest Population Density
correlation_top_density = top_population_density['Density(P/Km²)'].corr(top_population_density['PM2.5 AQI Value'])
print("Correlation between Population Density and PM2.5 AQI Value (Top 10 Density): ", correlation_top_density)

# Correlation for Top 10 Highest PM2.5 Levels
correlation_top_pm25 = top_pm25['Density(P/Km²)'].corr(top_pm25['PM2.5 AQI Value'])
print("Correlation between Population Density and PM2.5 AQI Value (Top 10 PM2.5): ", correlation_top_pm25)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Create the scatter plot
sns.scatterplot(x='Density(P/Km²)', y='PM2.5 AQI Value', data=top_population_density)

# Add annotations
for i in range(top_population_density.shape[0]):
    plt.text(top_population_density['Density(P/Km²)'].iloc[i], 
             top_population_density['PM2.5 AQI Value'].iloc[i], 
             top_population_density['Country'].iloc[i], 
             fontsize=9)

plt.title('Top 10 Highest Population Density vs PM2.5 AQI Value')
plt.xlabel('Population Density (people per km²)')
plt.ylabel('PM2.5 AQI Value')
plt.show()
```
